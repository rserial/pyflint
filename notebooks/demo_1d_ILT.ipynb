```python
data_filepath = Path('./example_data')
```

### T2

```python
filename_list_ET400us = [
    '230411-105923 T2Bulk (lyogel)',
    '230411-133146 T2Bulk (lyogel)',
    '230411-133451 T2Bulk (lyogel)'
]

filename_list_ET500us = [
    '230411-133753 T2Bulk (lyogel)',
    '230411-134057 T2Bulk (lyogel)',
    '230411-134400 T2Bulk (lyogel)'
]
```

Experiments for ET = 400 us

```python
filename_dir_list_ET400us = [
    # data_filepath / '230411-105923 T2Bulk (lyogel)' / 'processed_data' / 'T2Bulkdecay.dat',
    # data_filepath / '230411-133146 T2Bulk (lyogel)' / 'processed_data' / 'T2Bulkdecay.dat',
    data_filepath / '230411-133451 T2Bulk (lyogel)' / 'processed_data' / 'T2Bulkdecay.dat',
]
```

```python
average_signal = NMRsignal.average_signals(filename_dir_list_ET400us)

# ILT parameters setting
t1range = np.array([1e-4, 3e0])
t2range = None #1d experiment
dimKernel2D = [200, 1]
alpha = 5e-1
#Calling Flint method
flint = Flint(average_signal, dimKernel2D, "T2", alpha, t1range, t2range)
flint.solve_flint()
fig = flint.plot()
fig.show()
```

```python
for filename in filename_list_ET400us:

    #Define experiment filename
    T2Bulk_filename = filename

    #load NMR signal
    nmr_signal = NMRsignal.load_from_1d_txtfile(data_filepath/T2Bulk_filename/ 'processed_data'/'T2Bulkdecay.dat')
    # Adjusting time units to seconds
    nmr_signal.tau1 = nmr_signal.tau1

    # ILT parameters setting
    t1range = np.array([1e-4, 3e0])
    t2range = None #1d experiment
    dimKernel2D = [200, 1]
    alpha = 8e-1

    #Calling Flint method
    flint = Flint(nmr_signal, dimKernel2D, "T2", alpha, t1range, t2range)
    flint.solve_flint()
    fig = flint.plot()
    fig.show()
```

```python
for filename in filename_list_ET500us:

    #Define experiment filename
    T2Bulk_filename = filename

    #load NMR signal
    nmr_signal = NMRsignal.load_from_1d_txtfile(data_filepath/T2Bulk_filename/ 'processed_data'/'T2Bulkdecay.dat')
    # Adjusting time units to seconds
    nmr_signal.tau1 = nmr_signal.tau1

    # ILT parameters setting
    t1range = np.array([1e-4, 3e0])
    t2range = None #1d experiment
    dimKernel2D = [500, 1]
    alpha = 1

    #Calling Flint method
    flint = Flint(nmr_signal, dimKernel2D, "T2", alpha, t1range, t2range)
    flint.solve_flint()
    fig = flint.plot()
    fig.show()
```
