# Demo 2D data

```python
from pathlib import Path
import numpy as np
from pyflint.pyflint import FlintSignal, Flint
```

### Load data

For this example we will use 2d data from a T1(Inversion Recovery)T2 experiment.

```python
signal_2d = np.loadtxt(Path('./example_data') / "T1IRT2.dat", delimiter = " ")
t1_axis = np.loadtxt(Path('./example_data') / "T1IRT2_t1axis.dat")
t2_axis = np.loadtxt(Path('./example_data') / "T1IRT2_t2axis.dat")
```

### Perform ILT

Convert data to Flint signal format and initialise the Flint class. The ILT distribution is obtained by calling the `solve_flint` method. 

```python
signal = FlintSignal.load_from_data(signal_2d, t1_axis, t2_axis)
flint = Flint(signal, 
              kernel_shape = [32, 32], 
              kernel_name = "T1IRT2", 
              alpha = 1e-1, 
              t1range = [1e-4, 10e0], 
              t2range = [1e-4, 10e0], 
              )
flint.solve_flint()
flint.plot()
```
